# Read Raw Images

```python
import numpy as np
import subprocess
import glob
import skimage.io as imio
import re
from skimage.color import rgb2gray
import skimage
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_raw(fileName):
    
    rIm=np.fromfile(fileName,dtype=np.dtype('i2'))
    rows=2160
    runL=3840
    rIm=np.reshape(rIm[:rows*runL],(rows,runL))

    return rIm
```

```python
jIm=imio.imread("/home/qian/Desktop/qhGrab_20808_s01_00000.jpg")
jIm=skimage.transform.rotate(jIm,180)
rIm=read_raw("/home/qian/Desktop/nvcamtest_bayer_20808_00000.raw")

fig, ax = plt.subplots(1, 2, figsize=(20, 16))
ax[0].imshow(jIm)
ax[1].imshow(rIm)
plt.show()
```
